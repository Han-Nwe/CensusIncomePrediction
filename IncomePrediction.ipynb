# Data preparation

Preparation data for machine learning models involves a critical process of managing categorical data.

One-hot encoding is employed for nominal categorical data, where categories lack any inherent order. It transforms each category into binary columns with values of 0 or 1, representing the absence or presence of a particular category. This conversion guarantees that machine learning algorithms treat each category as distinct and avoids imposing any ordinal relationship between them.

```python
import pandas as pd

# Calculate the mode of the 'workclass' column
mode_workclass = my_df['workclass'].mode().iloc[0]

# Replace NaN values in the 'workclass' column with its mode
my_df['workclass'] = my_df['workclass'].fillna(mode_workclass)


# Calculate the mode of the 'workclass' column
mode_occupation = my_df['occupation'].mode().iloc[0]

# Replace NaN values in the 'workclass' column with its mode
my_df['occupation'] = my_df['occupation'].fillna(mode_occupation)


# Calculate the mode of the 'workclass' column
mode_native_country = my_df['native-country'].mode().iloc[0]

# Replace NaN values in the 'workclass' column with its mode
my_df['native-country'] = my_df['native-country'].fillna(mode_native_country)
```

# Optimisation for Logistic Regression

Optimization improves the performance of a Logistic Regression model by fine-tuning its hyperparameters, which are settings that control various aspects of the learning algorithm. Through optimization, we aim to find the best combination of hyperparameters that result in a model with better predictive accuracy and generalization to new, unseen data.

Logistic Regression models can sometimes overfit the training data, meaning they learn to perform exceptionally well on the training dataset but fail to generalize to new data. By optimizing hyperparameters such as the regularization strength (C parameter), you can prevent overfitting. Regularization introduces a penalty for large coefficient values, which encourages the model to be simpler and less prone to overfitting.

Optimization helps the model generalize better to unseen data by finding hyperparameters that balance between capturing complex relationships in the training data and avoiding fitting noise. This results in a model that is more likely to make accurate predictions on new, real-world examples.

Logistic Regression aims to estimate the probabilities of different classes. By optimizing hyperparameters, we can ensure that the model assigns appropriate probabilities to different classes, leading to better discrimination between classes and more accurate predictions.

Ultimately, the goal of optimization is to achieve a model that makes more accurate predictions on new, unseen data. By fine-tuning the hyperparameters, you're tailoring the model to the specific characteristics of your dataset, leading to better predictions and improved model performance.

In summary, optimization enhances the performance of a Logistic Regression model by guiding the model to a configuration of hyperparameters that reduces overfitting, improves generalization, speeds up convergence, and results in more accurate predictions. Through optimization, you can unlock the full potential of the Logistic Regression algorithm and achieve better results on real-world data.

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression


# Create a Logistic Regression model
logreg = LogisticRegression()

# Define hyperparameters and values to search
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10],
    'solver': ['lbfgs', 'liblinear', 'sag', 'saga']
}

# Perform grid search cross-validation
grid_search = GridSearchCV(logreg, param_grid, cv=5)
grid_search.fit(X_train, y_train)

# Get the best parameters and best model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

# Evaluate the best model on the test set
accuracy = best_model.score(X_test, y_test)
print(f"Best Parameters: {best_params}")
print(f"Test Accuracy: {accuracy}")
```

In the case of the Logistic Regression model, the optimization process led to the following best parameters such as C, 0.1, solver, sag. 

C parameter in Logistic Regression controls the strength of regularization. Regularization helps prevent overfitting by adding a penalty term to the loss function based on the magnitude of the coefficients. By optimizing the C parameter, you find the balance between fitting the training data well and not overfitting. Lower values of C increase regularization, which can help the model generalize better to new data.

The solver parameter determines the algorithm used for optimization during training. Different solvers use different optimization techniques to find the coefficients that minimize the loss function. In the opitimization, 'sag' is a variant of Stochastic Average Gradient descent. The choice of solver can impact convergence speed and effectiveness.

# Parameter of Optimization of Random Forest

For the optimization of a Random Forest model, you can tune several hyperparameters to improve its performance.

n_estimators: The number of decision trees in the forest. Increasing the number of trees can lead to a more robust model, but it may also increase training time. Too many trees can also lead to overfitting.

max_depth: The maximum depth of each decision tree. Increasing this parameter can lead to more complex models that may fit the training data better, but it can also increase the risk of overfitting.

min_samples_split: The minimum number of samples required to split an internal node. Increasing this parameter can make the model more robust against noise in the data.

min_samples_leaf: The minimum number of samples required to be at a leaf node. Similar to min_samples_split, increasing this parameter can prevent overfitting by ensuring a minimum amount of samples in each leaf.

max_features: The number of features to consider when looking for the best split. You can experiment with different values such as "auto", "sqrt", "log2", or an integer value to limit the number of features.

bootstrap: Whether to use bootstrapping (sampling with replacement) when building trees. Setting this parameter to False can help reduce overfitting.

criterion: The function to measure the quality of a split. Common options are "gini" for Gini impurity and "entropy" for information gain.

class_weight: This parameter can be used to address class imbalance by assigning different weights to different classes.

These parameters, along with others available in the RandomForestClassifier, can be optimized using techniques like Grid Search or Random Search, where you specify a range of values for each parameter and the optimization process tries different combinations to find the best set of parameters that maximize model performance on a validation set. It's important to strike a balance between model complexity and generalization to achieve the best results on unseen data.

```python
import warnings
warnings.filterwarnings('ignore')
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV

# Define the function to create the FNN model
def create_model(optimizer='adam', activation='relu', hidden_units=64):
    model = Sequential()
    model.add(Dense(hidden_units, input_dim=input_dim, activation=activation))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model

# Create the KerasClassifier wrapper
model = KerasClassifier(build_fn=create_model, epochs=10, batch_size=32, verbose=0)

# Define the parameter grid
param_grid = {
    'optimizer': ['adam', 'sgd', 'rmsprop'],
    'activation': ['relu', 'tanh'],
    'hidden_units': [32, 64, 128]
}

# Create a GridSearchCV object
grid = GridSearchCV(estimator=model, param_grid=param_grid, scoring='accuracy', cv=3)
grid_result = grid.fit(X_train, y_train)

# Print the best hyperparameters and best score
print("Best Parameters:", grid_result.best_params_)
print("Best Score:", grid_result.best_score_)
```
